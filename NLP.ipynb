```python
#program 1 tokenize text

def tokenization(text, method = ""):
  if method == "lines":
    text = text.split('\n')
    return text
  if method == "words":
    text = text.split()
    return text
  if method == "word_tokenize":
    text = word_tokenize(text)
    return text

text = """This is a sample text to perform word tokenizing
 This has multiple lines and lot of characters"""

print(tokenization(text, "lines"))
print(tokenization(text, "words"))
print(tokenization(text, "word_tokenize"))



```

```python
#Lemmetization and Stemming PorterStemmer() .stem

import nltk
nltk.download("punkt")
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

text = "Write written writting  writes"


words = word_tokenize(text)

stemmer = PorterStemmer()


stem_groups = {}


for word in words:
  stem  = stemmer.stem(word)
  if stem not in stem_groups:
    stem_groups[stem] = []
  stem_groups[stem].append(word)



for form, stem in stem_groups.items():
  print(f"{stem} -> {form}")
```

```python
# Smothing
import nltk

def laplace_smothing(words , vocab_size):
  smoothing = {}
  total_count = sum(words.values())
  for w , c in words.items():
    smoothing[w] = (c + 1) / (total_count + vocab_size)
  return smoothing



words = {"apple" : 3 , "banana" : 2 , "Oranage" : 2}
vocab_size = 100;

smooth = laplace_smothing(words , vocab_size)
for w , c in smooth.items():
  print(w  ,"=", c)
```

```python
#News Program

from sklearn.datasets import fetch_20newsgroups
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression


data = fetch_20newsgroups(subset  = "all")


X_train , X_test , y_train , y_test = train_test_split(data.data, data.target, test_size = 0.3,random_state= 42 )


model = make_pipeline(TfidfVectorizer(stop_words = "english") , LogisticRegression(max_iter = 1000))

model.fit(X_train , y_train)



accuracy = model.score(X_test, y_test)
print(accuracy)

text = "NASA is worst"

idx = model.predict([text])[0]

print("Prediction  = ", data.target_names[idx])




```

```python
#Disambiguity


import nltk

nltk.download("wordnet")

from nltk.corpus import wordnet as wn

def lexalgo(sentence , word):
  words = set(nltk.word_tokenize(sentence))
  synsets = wn.synsets(word)


  if not synsets:
    return None

  def getoverlap(sense):
    return len(words & set(nltk.word_tokenize(sense.definition())))


  return max(synsets,key = getoverlap)










```

```python
result = lexalgo("THe dog is sitting near the bank", "bank")
print(result.definition())
```

```python
#Question answer system


from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity


corpus = [
    "newton invented laws of motion ",
    "AryaBhata invented Zero " ,
    "ISRO was started BY APJ Abdul Kalam"
]


questions = ["who started ISRO","who invented Zero" ]


vec = TfidfVectorizer(stop_words = "english")

X = vec.fit_transform(corpus)


for q in questions:
  q_vec= vec.transform([q])
  idx = cosine_similarity(q_vec , X).argmax()
  print("Q = ",q)
  print("A = ",corpus[idx])




```

```python
#positional Encoding


import numpy as np

def getpe(dim , seq_len):
 pos = np.arange(seq_len)[:,None]

 div = 10000**(np.arange(0,dim,2)/dim)

 pe = np.zeros((seq_len , dim))

 pe[:,0::2] = np.sin(pos/div)
 pe[:,1::2] = np.cos(pos/div)
 return pe






pe = getpe(512,10)
print(pe[:2,:4])

```

```python
from collections import Counter
```

```python
#program Good Turing

def getgood(counts):
  freq_counts = Counter(counts.values())
  total = sum(counts.values())

  good_grps = {}
  for word , count in counts.items():
    nk = freq_counts[count]
    nk1 = freq_counts.get(count +1 , 0)

    if nk1 > 0:
      adjusted = (count + 1) * (nk1 / nk)
    else:
      adjusted = count


    good_grps[word] = adjusted / total

  return good_grps













ngram_counts = {
    'the': 5,
    'of': 3,
    'and': 3,
    'to': 2,
    'a': 2,
    'in': 1
}


res = getgood(ngram_counts)
print(res)



```
